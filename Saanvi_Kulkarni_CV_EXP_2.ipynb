Erosion

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread('/content/image1.jpg', cv2.IMREAD_GRAYSCALE)
cv2_imshow(image)
```

```python
# Define a structuring element (kernel)
kernel = np.ones((5, 5), np.uint8)

# Perform erosion
eroded_image = cv2.erode(image, kernel, iterations=8)
```

```python
plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1, 2, 2), plt.imshow(eroded_image, cmap='gray'), plt.title('Eroded')
plt.show()
```

Remove Noise

```python
# Read the image
image = cv2.imread('/content/image2.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply GaussianBlur to reduce noise
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
```

```python
# Define structuring elements
kernel_opening = np.ones((6, 6), np.uint8)
kernel_closing = np.ones((2, 2), np.uint8)
```

```python
# Perform morphological opening to remove noise
opened_image = cv2.morphologyEx(blurred_image, cv2.MORPH_OPEN, kernel_opening)

# Perform morphological closing to fill holes or gaps
closed_image = cv2.morphologyEx(opened_image, cv2.MORPH_CLOSE, kernel_closing)
```

```python
# Display the original, opened, and closed images
plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1, 3, 2), plt.imshow(opened_image, cmap='gray'), plt.title('Opened')
plt.subplot(1, 3, 3), plt.imshow(closed_image, cmap='gray'), plt.title('Closed')
plt.show()
```

Gray scale 3x3 dilation and erosion

```python
# Read the image
image = cv2.imread('/content/image3.png', cv2.IMREAD_GRAYSCALE)
```

```python
# Define a 3x3 square structuring element (kernel)
kernel = np.ones((3, 3), np.uint8)

# Perform gray-scale dilation
dilated_image = cv2.dilate(image, kernel, iterations=1)

# Perform gray-scale erosion
eroded_image = cv2.erode(image, kernel, iterations=1)

# Compute morphological gradient
gradient_image = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)

# 2nd method to compute morphological gradient
gradient_image2 = dilated_image - eroded_image
cv2_imshow(gradient_image2)
```

```python
# Display the original, dilated, eroded, and gradient images
plt.subplot(1, 4, 1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1, 4, 2), plt.imshow(dilated_image, cmap='gray'), plt.title('Dilated')
plt.subplot(1, 4, 3), plt.imshow(eroded_image, cmap='gray'), plt.title('Eroded')
plt.subplot(1, 4, 4), plt.imshow(gradient_image, cmap='gray'), plt.title('Morphological Gradient')
plt.show()
```
